```python
from pyspark.sql.functions import col

# Assuming df_clean is your DataFrame
df_clean = df_clean.withColumn("monthly_salary", col("clean_salary") / 12)

# Show the DataFrame with the new column
df_clean.show()
```

```python
from pyspark.sql.functions import col, sum as _sum
# Group by gender and aggregate the sum of "clean_salary"
df_gender_summary = df_clean.groupBy("gender").agg(sum("clean_salary").alias("summary"))

# Show the resulting DataFrame
df_gender_summary.show()
```

```python
df1 = df_clean.groupBy('gender').agg( sum('clean_salary').alias('summary'),
avg('clean_salary').alias('mean'),
min('clean_salary').alias('min'),
max('clean_salary').alias('max'))
```

```python
df_clean = df_clean.withColumn("clean_salary", col("clean_salary").cast("double"))
df_city_avg_salary = df_clean.groupBy("City").agg(avg("clean_salary").alias("average_salary"))
highest_avg_salary_city = df_city_avg_salary.orderBy(col("average_salary").desc()).limit(1)
highest_avg_salary_city.show()
```

```python
df_avg_salary = df_clean.groupBy("JobTitle", "gender").agg(avg("clean_salary").alias("average_salary"))
df_pivot = df_avg_salary.groupBy("JobTitle").pivot("gender").agg(avg("average_salary"))
df_residual = df_pivot.withColumn("residual", col("Female") - col("Male"))
df_residual.show()
```
